# Integrating Datasets

## Import cleaned individual datasets

- Dataset 1: Bulls Season Statistics Data
- Dataset 2: Bulls Draft Data

Both datasets were previously cleaned and pushed to our data folder.

```python
import pandas as pd
```

```python
df1 = pd.read_csv("../data/processed/bulls_season_data.csv")
df1.head()
```

```python
df2 = pd.read_csv("../data/processed/draft_data_cleaned.csv")
df2.head()
```

### Merge Datasets 
---

- Left: "Season"
- Right: "Year"

Both columns hold the same information, that being the year of that draft/season. We inner join the two datasets

```python
# Merge the two DataFrames on the 'season'  column for df1, and year for df2
merged_df = pd.merge(df1, df2, left_on='Season', right_on='Year', how='inner')
merged_df.head()
```

We check the columns to ensure all data was correctly integrated

```python
merged_df.columns
```

### Missing Values
---

We look at the missing values in the merged dataset to also ensure data was properly integrated. There are columns in which missing values are intentional (playoffs, awards).

```python
merged_df.isnull().sum()
```

We see that there are four missing values across Age, Pos, G; mostly columns from the draft dataset. We'll explore what observations are missing

```python
merged_df[merged_df['Age'].isnull()]["Player"]
```

Let's check the draft dataset to see if these were missing in the second dataset

```python
# check if missing ages are missing in dataset "df2"
df2[df2["Player"].isin(merged_df[merged_df['Age'].isnull()]["Player"])]
```

We find that these values are missing. When we check ESPN for each of their stats, the reason these are missing is because they did not play in their rookie season. This should not take away from "Age" or "Position" so we will have to manually fill these out in our merged df.

```python
merged_df["Pos"].unique()
```

```python
merged_df.loc[merged_df['Player'] == 'Marko Simonovic', 'Age'] = 21
merged_df.loc[merged_df['Player'] == 'Marko Simonovic', 'Pos'] = 'C'

merged_df.loc[merged_df['Player'] == 'JamesOn Curry', 'Age'] = 22
merged_df.loc[merged_df['Player'] == 'JamesOn Curry', 'Pos'] = 'PG'

merged_df.loc[merged_df['Player'] == 'Mario Austin', 'Age'] = 21
merged_df.loc[merged_df['Player'] == 'Mario Austin', 'Pos'] = 'SF'

merged_df.loc[merged_df['Player'] == 'Tommy Smith', 'Age'] = 23
merged_df.loc[merged_df['Player'] == 'Tommy Smith', 'Pos'] = 'SF'
```

For the rest of the player stats, if missing, we can fill in with a zero. 

```python
cols_to_fill = ['G', 'GS','MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%',
       'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK',
       'TOV', 'PF', 'PTS']
merged_df[cols_to_fill] = merged_df[cols_to_fill].fillna(0)
```

```python
merged_df.head()
```

Finally, we can replace missing values in "Playoffs" and "Awards" with a simple value like "None" to indicate that there was no playoff appearance or awards won for that player.

```python
# check unique values for playoffs
merged_df["Playoffs"].unique()
```

```python
merged_df["Playoffs"] = merged_df["Playoffs"].fillna("Missed Playoffs")
```

```python
# check unique values for awards
merged_df["Awards"].unique()
```

If we want to gain more information about player awards, we can create a column for "ROY" whith the numeric value of their ranking in voting for the element. We have one player that won "sixth man of the year" or "6MOY" in the code, but since we only have one occurence we wouldn't have enough evidence to suggest the effect of winning sixth man of the year.

```python
def extract_roy_rank(awards):
    if pd.isnull(awards):
        return 0
    # extract the first 5 characters of the awards string
    awards_prefix = awards[:5]
    # extract the last character of the awards string
    awards_suffix = awards_prefix[-1]
    # place suffix into column "ROY"
    if awards_prefix.startswith("ROY-") and awards_suffix.isdigit():
        return int(awards_suffix)
    else:        
        return 0

merged_df["ROY"] = merged_df["Awards"].apply(extract_roy_rank)
merged_df
```

```python
# ensure that if awards contains "ROY", ROY column contains correct rank
merged_df[merged_df["Awards"].str.contains("ROY", na=False)][["Player", "Awards", "ROY"]]
```

Now we can fill the awards column, if Nan, with "None"

```python
merged_df["Awards"] = merged_df["Awards"].fillna("None")
```

```python
merged_df
```

We now have a fully cleaned merged datset. To reiterate the steps we took:

- We imported both Chicago Bulls season & draft datasets
- We inner joined the datasets on the "season/year" column
- We cleaned missing values
    - We hard-coded missing Age/Pos for players that did not play their rookie season
    - We replaced missing values with zeros for those players
    - We filled missing playoffs/awards with "Missed" or "None"
- We extracted Rookie Of The Year ranking for the player

We can finally upload this dataset to the data folder in github.

```python
write_csv_path = "../data/processed/final_bulls_data.csv"
merged_df.to_csv(write_csv_path, index=False)
```
