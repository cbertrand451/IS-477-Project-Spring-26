```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/bulls_data.csv")
df.head()
```

```python
# convert season to integer, removing hyphen-"next year" part

df["Season"] = df["Season"].apply(lambda x: int(x.split("-")[0]))
df.head()
```

```python
# drop "Unnamed: 8", "Unnamed: 15" column, which is empty
df.drop(columns=["Unnamed: 8", "Unnamed: 15"], inplace=True)
df.head()

# despite NAs in the Playoffs column, we leave them be to show a "missed" playoffs season.
```

```python
# Convert "Finish" to integer by keeping the first letter of string and converting to integer
df["Finish"] = df["Finish"].apply(lambda x: int(x[0]) if isinstance(x, str) else np.nan)
df.head()
```

```python
# drop Lg column, which is always "NBA"
df.drop(columns=["Lg"], inplace=True)
df.head()
```

```python
# rewrite the csv file with the cleaned data
df.to_csv("data/bulls_dataset2.csv", index=False)
```
